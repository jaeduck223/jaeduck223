```python
data = pd.DataFrame({'k1': ['one', 'two'] * 3 + ['two'],
                    'k2': [1,1,2,3,3,4,4]})
data
```

```python
data.duplicated() #duplicated 메소드는 각 row가 중복인지 아닌지를 알려주는 Series를 반환한다.(중복된 것이 참.)
```

```python
data.drop_duplicates()
#drop_duplicates는 duplicated 배열이 False인 DataFrame을 반환한다.(k1, k2 컬럼 둘다 따져서 중복 된 것은 제외함.)
```

```python
#새로운 컬럼 하나를 추가하고 'k1'컬럼에 기반해 중복을 걸러낸다.
data['v1'] = range(7)
data
```

```python
data.duplicated()
```

```python
data.drop_duplicates(['k1']) #k1만 고려하여서 중복된 것을 다 날려버리고, 남은 것을 출력.
```

```python
#duplicated와 drop_duplicates는 기본적으로 처음 발견된 값을 유지한다.
data.drop_duplicates(['k1', 'k2']) #k1, k2 컬럼 둘 다 고려함.
```

```python
#duplicated와 drop_duplicates는 기본적으로 처음 발견된 값을 유지한다.
#keep='last'옵션을 넘기면 마지막으로 발견된 값을 반환한다.(중복된 것 중 마지막 것은 남기고, 중복된 것들을 날리고 출려함.)
data.drop_duplicates(['k1', 'k2'], keep='last')
```

```python
data1 = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon',
                             'Pastrami', 'corned beef', 'Bacon',
                            'pastrami', 'honey ham', 'nova lox'],
                    'ounces': [4, 3, 12, 6, 7.5, 8, 3, 5, 6]})
data1
```

```python
#해당 육류가 어떤 동물의 고기인지 알려주는 컬럼을 추가한다고 가정하자.
meat_to_animal = {
    'bacon': 'pig',
    'pulled pork': 'pig',
    'pastrami': 'cow',
    'corned beef': 'cow',
    'honey ham': 'pig',
    'nova lox': 'salmon'
}
```

```python
lowercased = data1['food'].str.lower() #데이터에 대소문자 섞여있으므로, 소문자로 변경(고급 스킬)
lowercased
```

```python
#Series의 map 메소드는 사전류의 객체나 어떤 함수를 받을 수 있다.
data1['animal'] = lowercased.map(meat_to_animal) #map 메소드 이용
data1
```

```python
data1['food'].map(lambda x: meat_to_animal[x.lower()]) #함수를 넘겨서 같은 일을 수행 가능
```

```python
data3 = pd.DataFrame(np.arange(12).reshape((3,4)),
                   index = ['Ohio', 'Colorado', 'New York'],
                   columns = ['one', 'two', 'three', 'four'])
data3
```

```python
transform = lambda x: x[:4].upper() #소문자를 대문자로 치환함.

data3.index.map(transform)
```

```python
data3
```

```python
data3.index = data3.index.map(transform)
data3
```

```python
data3.rename(index=str.title, columns=str.upper) #원래 객체를 변경하지 않고 새로운 객체를 생성하려면 rename 메소드를 사용한다.
#Series.str.lower : Converts all characters to lowercase.
#Series.str.upper : Converts all characters to uppercase.
#Series.str.title : Converts first character of each word to uppercase and remaining to lowercase.
#Series.str.capitalize :Converts first character to uppercase and remaining to lowercase.
#Series.str.swapcase :Converts uppercase to lowercase and lowercase to uppercase.
```

```python
data3.rename(index={'OHIO':"INDIANA"},
           columns = {'three':'peekaboo'})
```

```python
data3
```

```python
#rename 메소드를 사용하면 DataFrame을 직접 복사해서 index와 column 속성을 갱신할 필요없이 바로 변경 가능
#원본 데이터를 바로 변경하려면, inplace=True 옵션을 사용하면 된다.

data3.rename(index={'OHIO': 'INDIANA'}, inplace=True)
data3
```
